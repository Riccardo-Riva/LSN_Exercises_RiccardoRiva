Terminata la fase di equilibrazione si può procedere con la valutazione delle autocorrelazioni e la stima della larghezza dei blocchi.

Per tutte e tre le fasi vengono generate $5\cdot10^5$ configurazioni tramite l'algoritmo di Metropolis, per ciascuna configurazione vnegono caloclati i valori istantanei di $V/N$ e $P$  e innanzitutto si musura la  funzione di autocorrelazione di queste quantità.
I file contenenti i valori istantanei misurati sono salvati nelle directory **solid**, **liquid** e **gas**. 
La funzione di autocorrelazione viene valutata come segue:

$$
Ac_{X}(n) = \frac{<X(n')X(n'+ n)> - <X(n)X(n'+n)>}{\sigma^2_X}
$$

L'autocorrelazione è quindi intesa come funzione del numero di step $n$, che rappresentazno la differenza di step fra i due valori di $X(m)$ fra i quali si valuta la correlazione.
A livello numerico l'autocorrelazione viene valutata considerando i $n_{max}=5 \cdot 10^5$ valori istantanei misrati e calcolando:

$$
Ac_{[X]}(n) = \frac{\frac{1}{n_{max}-n} \sum^{n_{max}-n}_{n'=0}X(n')X(n'+n) - \frac{1}{n_{max}-n} \sum^{n_{max}-n}_{n'=0}X(n')\frac{1}{n_{max}-n} \sum^{n_{max}-n}_{n'=0}X(n'+n)}{\frac{1}{n_{max}} \sum^{n_{max}}_{n'=0}X^2(n')-(\frac{1}{n_{max}} \sum^{n_{max}}_{n'=0}X(n'))^2}
$$

Per l'analisi dell'autocorrelazione sono stati fatti dei grafici con $n$ che assume valori da $0$ a $1000$ (con salti di $5$) e da $0$ a $5000$ (con salti di $20$), per tutte e tre le fasi. I grafici sono salvati nelle directory **solid**, **liquid** e **gas**.  Di seguito sono mostrati i grafici con $n$ che arriva fino a $1000$

### Esercizio 07.4




Ora possiamo svolgere la simulazione vera e propria. Vengono simulati $20$ blocchi da $5000$ step ciascuno per un totale di $10^5$ step Monte Carlo. Vengono misurati i valori di $U/N$ e $P$ per poi produrre un grafico dell'andamento delle stime di queste grandezze e delle loro incertezze all'aumentare del numero di blocchi. Dopodichè viene prodotto un grafico di confronto fra la $g(r)$ valutata con la simulazione Monte Carlo e quella valutata con la Dinamica Molecolare, insieme alle rispettive incertezze.

### Fase Solida

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy import constants as const
from scipy.constants import physical_constants
from math import sqrt

######## ARGON ########

sigma = 0.34e-9
T0 = 120
m = 39.948 * physical_constants['atomic mass constant'][0]
epsilon = T0 * const.k
time0 = sigma * sqrt(m / epsilon)
P0 = epsilon / (sigma**3)

#######################


plt.rc('text', usetex=True)

v, errv = np.loadtxt("MonteCarlo_NVT/solid/output.epot.0", usecols = (2,3), delimiter=' ', unpack ='true') * epsilon
p, errp = np.loadtxt("MonteCarlo_NVT/solid/output.pres.0", usecols = (2,3), delimiter=' ', unpack ='true') * P0
r, gMC, errMC = np.loadtxt("MonteCarlo_NVT/solid/output.gave.0", usecols = (0,1,2), delimiter=' ', unpack ='true') 
gMD, errMD = np.loadtxt("MonteCarlo_NVT/solid/output.gaveMD.0", usecols = (0,1), delimiter=' ', unpack ='true') 


x = np.arange(len(v)) + 1 
r = r*sigma

plt.figure(figsize=(9,6))
plt.errorbar(x, v, yerr=errv, color='green')
plt.xlabel('Number of Blocks', fontsize=18)
plt.ylabel('V/N [J]', fontsize=18)
plt.title('Potenzial Energy - Solid', fontsize=22)
plt.grid(True)

plt.figure(figsize=(9,6))
plt.errorbar(x, p, yerr=errp)
plt.xlabel('Number of Blocks', fontsize=18)
plt.ylabel('P [Pa]',fontsize=18)
plt.title('Pressure - Solid', fontsize=22)
plt.grid(True)

plt.figure(figsize=(9,6))
plt.errorbar(r, gMC, yerr=errMC, color='blue', label='Monte Carlo')
plt.errorbar(r, gMD, yerr=errMD, color='red', label='Molecular Dynamics')
plt.xlabel('r [m]', fontsize=18)
plt.ylabel(r'$g(r)$',fontsize=18)
plt.title('Radial Distribution - Solid', fontsize=22)
plt.legend()
plt.grid(True)

plt.show()
```
