 ### Esercizio 01.1
 In questo esercizio stimiamo il valore dell'integrale: $\langle r \rangle = \int_0^1 r dr$, usando metodi Monte Carlo. Per questo calcolo utilizziamo il generatore di Numeri Pseudo-Casuali fornito durante le lezioni del *Laboriatorio di Simulazione Numerica*.
In particolare vogliamo mostrare (tramite opportuni grafici) come la stima fornita dai metodi Monte Carlo migliori all'aumentare del numero di "test" che si eseguono.
Il codice per questo esercizio Ã¨ quello scritto nel file *Exercise01.1.cpp*. Inoltre in tutto l'esercizio si sfruttano le funzione scritte nel file *functions.cpp* che permettono di eseguire una serie di operazioni con i **vector** della STL.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rc('text', usetex=True)
plt.rc('font', family='arial')

val, err = np.loadtxt("montecarlo.dat", usecols=(0,1), delimiter=' ', unpack='true')
x = np.arange(len(val)) + 2

plt.figure(figsize=(10,7))
plt.errorbar(x,val-0.5,yerr=err)
plt.xlabel(r'Number of Blocks', fontsize=20)
plt.ylabel(r'$<r>-\frac{1}{2}$', fontsize=20)
plt.grid(True)
plt.title('Mean', fontsize=20)

plt.show()
```

Nel grafico Ã¨ mostrato l'andamento del valore stimato per $\langle r \rangle$ e della sua incertezza all'aumentare del numero di lanci. In particolare sono stati effettuati 1 milione di lanci, divisi in 100 blocchi da 10000. Abbiamo calcolato la media in ogni blocco per poter valutare l'incertezza con il metodo del data blocking.

Per questa stima abbiamo appunto fatto uso di un Generatore di numeri Pseudo-Casuali, in particolare generando numeri distribuiti uniformemente fra 0 e 1. Ci interessa poi calcolare
anche la varianza dei nostri numeri e verificare che sia approssimativamente $\frac{1}{12}$, cioÃ¨ il valore esatto per tale distribuzione.

```python
val, err = np.loadtxt("montecarlo_variance.dat", usecols=(0,1), delimiter=' ', unpack='true')
x = np.arange(len(val)) + 2

plt.figure(figsize=(10,7))
plt.errorbar(x,val-1/12,yerr=err)
plt.xlabel(r'Number of Blocks', fontsize=20)
plt.ylabel(r'$\sigma^2 - \frac{1}{12}$', fontsize=20)
plt.grid(True)
plt.title('Variance', fontsize=20)

plt.show()
```

### Esercizio 01.3
L'obiettivo di questo ultimo esercizio Ã¨ quello di simulare l'esperimento di Buffon. L'esperimento effettivo coinvolge una griglia con linee parallele a una distanza $d$ l'una dall'altra, e un ago di lunghezza $l$. Se consideriamo il punto medio dell'ago, questo cadrÃ  a una distanza $\leq \frac{d}{2}$ da una delle linee, in particolare la posizone $x$ (lungo la direzione ortogonale alla linea) del punto medio rispetto a questa linea Ã¨ una variabile casuale che possiamo considerare distribuita uniformemente fra $-\frac{d}{2}$ e $+\frac{d}{2}$. L'angolo che l'ago forma rispetto alla linea Ã¨ chiaramente una variabile casuale distribuita fra $0$ e $2\pi$. Per cui la simulazione avviene campionando questa coppia di variabili un certo numero di volte e contando quante volte l'ago interseca la linea, cioÃ¨ quante volte l'intervallo $[x-\frac{1}{2}l|sin(\theta)|,\:x+\frac{1}{2}l|sin(\theta)|]$ contiene lo zero. Per campionare il valore dell'angolo senza usare un valore di $\pi$ noto a priori si usa il metodo presentato nella Lezione 2 del corso, generando un punto in un quadrato centrato nell'origine e calcolando direntamente il seno dell'angolo fra il segmento che unisce il punto con l'origine e l'asse x.

In maniera simile al primo esercizio vengono generati $10^6$ campioni divivisi in $100$ blocchi da $10^4$ lanci. In ciascuno blocco si conta quante volte l'ago interseca la linea e si da una stima di $\pi$ tramite la formula $$\pi = \frac{2LN_{thr}}{N_{hit}d}$$
dopodiche si esegue la media su i vari blocchi anche per stimare l'incertezza con la deviazione standard della media.

```python
val, err = np.loadtxt("pi.dat", usecols=(0,1), delimiter=' ', unpack='true')
x = np.arange(len(val)) + 2

plt.figure(figsize=(9,6))
plt.errorbar(x,val-3.1415926535,yerr=err)
plt.xlabel(r'Number of Blocks', fontsize=20)
plt.ylabel(r'$<\pi>-\pi_{real}$', fontsize=20)
plt.grid(True)
plt.title('Mean', fontsize=20)

plt.show()

print("Estimate:")
print(val[-1])
print("Uncertainty:")
print(err[-1])
```
