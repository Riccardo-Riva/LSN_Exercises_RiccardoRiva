### Esercizio 09.1
In questo esercizio vogliamo implementare un Algoritmo Genetico per risolvere il Problema del Commesso Viaggiatore (TSP). Dapprima risolviamo il problema nel caso di città poste su una circonferenza di raggio unitario, dopodichè consideriamo le citta poste in un quadrato di lato unitario.

Prima di tutto bisogna generare le posizioni delle città nei due casi. Questo viene codificato nel file *Cities.cpp* in cui tramite il generatore di numeri casuali fornito a lezione vengono generate 32 posizioni di città in entrambi i casi, che vengono salvate nei file *cities_circumference.dat* e *cities_plane.dat*.

Per implementare la popolazione di individui vengono create le classi **Individual** e **Population**, dichiarate nei file *individual.h* e *population.h*.

La classe **Individual** è quella che contiene le informazioni sul signolo individuo. Infatti ha come membro il **vector** di interi **m_path**, destinato a contenere la successione di città seguita dal singolo individuo (l'indice intero che indica una citta viene sostanzialmente definito nel **main** durante l'esecuzione).
Tutti i metodi definiti (oltre a costruttori, distruttori, etc.) hanno o il compito di modificare il contenuto del **vector** o il compito di accedere in qualche modo ai dati memorizzati. Noi durante tutto il processo di ottimizzazione stocastica vogliamo che ciascun individuo abbia sempre nella prima posizione di **m_path** l'indice relativo alla prima città (cioè il numero 0), in modo da diminuire la degenerazione dei percorsi equivalenti. 
Per questo viene definito il metodo **Bond**, che viene utilizzato dopo ogni modifica con il compito di fare uno switch degli elemnti del vector fino ad avere l'elemento 0 in prima posizione.
I metodi principali della classe sono quelli che implementano gli operatori di Mutazione dell'algoritmo genetico.

- **Swap:** scambia due elementi del **vector**
- **Cycle:** prende una porzione continua del **vector** ed esegue una permutazione ciclica su quest'ultima
- **Invert:** prende una porzione continua del **vector** e la inverte
- **SwapBlocks** scambia due porzioni del **vector** con lo stesso numero di elementi
- **TripleSwap:** sceglie tre elementi in posizioni casuali del **vector** e li scambia in modo ciclico.

Un altro metodo importante è il metodo **ChangeValue** che cambia direttamente il valore di un elemento del vector. Questo è il metodo più delicato perchè quando viene usato non garantisce che l'individuo alla fine rispetti tutti i vincoli, soprattuto il fatto che tutti gli elementi devono essere diversi (perchè si passa una sola volta da ciascuna città). Per questo è anche definito un metodo **Check**, da usare quando viene usato anche **ChangeValue**, che verifica che l'individuo rispetti tutti i vincoli imposti.
Nel file *individual.h* è dichiarata anche la funzione **indexPBC** che serve per avere degli indici sempre nel range del **vector** (da 0 a N-1, dove N è la dimensione dell'eventuale vettore).


La classe **Population** contiene le informazioni sull'intera popolazione di individui e soprattutto contiene i metodi per eseguire l'evoluzone dell'algoritmo genetico.
Innanzitutto si ha un **vector** di **Individual**: **m_population**. Questa è effettivamente la popolazione di individui su cui si vuole lavorare. Si ha poi un puntatore alla classe **Random** per poter utilizzare il generatore di numeri casuali che verrà evenetualmente dichiarato nel *main*.
Si ha poi la matrice (costruita tramite i **vector**) **m_distances** che contiene le distanze fra le varie città (nel senso che **m_distances[i][j]** è la distanza fra la città *i* e la città *j*). Questa matrice viene utilizzata dal metodo **Cost** per misurare il costo di un individuo della popolazione, cioè la sua lunghezza.
Si hanno poi metodi per caricare i dati di una popolazione da file e anche per caricarli su un file, oltre che a caricare gli elementi della matrice delle distanze.
Il metodo **Sorting()** ha il compito di ordinare gli individui della popolazione in ordine crescente di costo. Questo algoritmo non va a modificare direttamente il **vector** della popolazione ma modifica il vector **sortedIndex** che alla fine contiene gli indici del vettore di popolazione in ordine crescente di lunghezza (**sortedIndex[0]** contiene l'indice all'individuo con costo minore di **m_population**). L'algoritmo di sorting utilizzato è il **merge sort**.
Il metodo **Generate** è quello che effettivamente implementa l'Algoritmo Genetico.

Vengono selezionati due genitori. Gli individui sono già stati ordinati col metodo **Sorting** in un passaggio precedente e la selezione avviene generando un numero casuale fra $0$ e $1$ e calcolando $i=\lfloor N a^2 \rfloor$, dove $\lfloor \: \rfloor$ indica la parte intera e N è il numero di individui della popolazione. Dopodichè si seleziona l'i-esimo individuo della popolazione in ordine di costo.
Selezionati due individui si ha una probabilità dell'$85\%$ che effettuino un Crossover, nel modo indicato a lezione.
Infine si hanno delle mutazioni con le seguenti probabilità:

- Swap: p=$1\%$
- Cycle: p=$1\%$
- Invert: p=$1\%$
- SwapBlocks: p=$1\%$
- Triple Swap: p=$1\%$

I vari indici degli elemnti da scambiare o le dimensioni dei blocchi da modificare in tutti e quattro i casi vengono scelti in modo casuale (tra valori compatibili con la mutazione che si stà effettuando).

### Evoluzione
Innanzitutto dobbiamo generare la popolazione iniziale. Questo viene fatto nel programma scritto nel file *FirstGen.cpp* che stampa la generazione iniziale sul file *first_generation.dat*. Per avere delle permutazioni casuali dei numeri da $0$ a $31$ venogno eseguiti una serie di scambi casuali. Il numero di scambi casuali è anch'esso un numero casuale fra 2 e 10 volte il numero di città.
La stessa popolzione iniziale viene usata per entrambi i casi.
Il file *TSP.cpp* è il main principale in cui viene eseguita l'evoluzione per entrambi i casi. Venogno lette tutte le informazioni necessarie dai file, vengono calcolate le varie distanze e si procede con l'Evoluzione. Per controllare la convergenza dell'algoritmo, a ogni generazione si stampa la lunghezza del miglior percorso. Nel caso della circonferenza si è trovato che $1000$ generazioni sono sufficienti perchè la miglior lunghezza si fissi su un certo valore.
Il problema nel piano invece è più complicato e in questo caso sono necessarie $8000$ generazioni per arrivare a convergenza.

Di seguito sono riportati i grafici dell'andamento della media dei valori di lunghezza del percorso sulla miglior metà della popolazione.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rc('text', usetex=True)

Gen1, Circ = np.loadtxt("TSP/length_averages_circ.dat", usecols = (0,1), delimiter=' ', unpack ='true')
Gen2, Plane = np.loadtxt("TSP/length_averages_plane.dat", usecols = (0,1), delimiter=' ', unpack ='true')

plt.figure(figsize=(9,6))
plt.plot(Gen1, Circ, color='red')
plt.xlabel('Generation', fontsize=18)
plt.ylabel('$Average$', fontsize=18)
plt.title('Averages of Length - Circumference', fontsize=22)
plt.grid(True)
plt.show()

plt.figure(figsize=(9,6))
plt.plot(Gen2, Plane, color='red')
plt.xlabel('Generation', fontsize=18)
plt.ylabel('$Average$', fontsize=18)
plt.title('Averages of Length - Plane', fontsize=22)
plt.grid(True)
plt.show()
```
