### Esercizio 06.1
In questa esercitazione vogliamo simulare il modello di Ising unidimensionale, confrontando poi i risultati con la soluzione analitica. Simuliamo un sistema unidimensionale di $N = 50$ momenti magnetici a differenti temperature per andare a misurare a ogni temperatura i valori di $U(N,T)/N$ (Energia Interna per Spin), $C(N,T)$ (Capacità termica), $M(N,T)/N$ (Magnetizzazione media), $\chi(N,T)$ (Suscettibilità Magnetica). 
In particolare come metodi di simulazione usiamo sia il campionamento con l'algoritmo di Metropolis, sia il campionamento di Gibbs, per verificare la funzinalità di entrambi i metodi.

Il codice fornito dal professore durante le lezioni è stato cmpletato in maniera che si possa scegliere fra esegure il Campionamento di Gibbs o di Metropolis e che a ogni step della simulzione si esegua la misura di tutte le grandezze dette sopra.

L'Hamiltoniana del sistema di Spin, che descrive sia l'interazione di ogni spin con i due immediatamente a fianco, sia l'interazione con un campo magnetico esterno, è data da:

$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
- h\sum_{i=1}^N s_i 
$$

Nelle nostre simulazioni abbiamo $J=1,00$ e $h=0,00$ o $h=0,02$.
Le temperature che si vogliono simulare sono quelle comprese fra 0,5 e 2,0 (in unità in cui $k_B=1$, $\mu_B=1$ e $\hbar=1$).

## Equilibrazione

Il test viene eseguito alle temperature $0,5$, $1,0$, $1,5$ e $2,0$, sia per il campionamento di Gibbs che per quello di Metropolis. In tutti questi test la configurazione iniziale è quella in cui tutti gli spin hanno la stessa probabilità di avere valori $+1$ o $-1$, che corrisponde a una temperatura $T=\infty$.

## Simulazione
Per quanto riguarda la simulazione vera e propria, l'intervallo di di temperature scelto (da $0,5$ a $2,0$) viene diviso in $100$ sottointervalli identici, avendo così 101 valori di temperatura da simulare corrispondenti agli estremi di questi sottointervalli. In totale eseguiamo $4$ simulazioni, per ognuno dei due metodi di campionamento simuliamo il caso in cui $h=0,00$ e il caso in cui $h=0,02$. I valori $U(N,T)/N$, $C(N,T)$, $M(N,T)/N$, $\chi(N,T)$ vengono stimati come media su tutti gli step della simulazione e l'errore viene stimato tramite il metodo del *data blocking*. Le simulazioni a ogni temperatura consistono di $30$ blocchi da $2 \cdot 10^4$ step ciascuno, per un totale di $6 \cdot 10^5$ step per ogni simulazione. I risultati di tutte le simulazioni sono salvati nelle directory **Metro_H_0,00**, **Gibbs_H_0,00**, **Metro_H_0,02** e **Gibbs_H_0,02**.
I grafici sottostanti mostrano l'andamento al variare della temperatura dei valori di $U(N,T)/N$, $C(N,T)$, $\chi(N,T)$ nel caso $h=0,00$ e del valore di $M(N,T)/N$ nel caso $h=0,02$. In ogni grafico sono presenti i valori di entrambi i metodi di campionamento con le relative incertezze e il grafico della soluzione analitica.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rc('text', usetex=True)

T0 = 0.5
T1 = 2.0
points=101
T = np.linspace(T0,T1,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )

um, errum = np.loadtxt("ISING_1D/Metro_H_0.00/energy.out", usecols = (0,1), delimiter=' ', unpack ='true')
ug, errug = np.loadtxt("ISING_1D/Gibbs_H_0.00/energy.out", usecols = (0,1), delimiter=' ', unpack ='true')
cm, errcm = np.loadtxt("ISING_1D/Metro_H_0.00/heat_capacity.out", usecols = (0,1), delimiter=' ', unpack ='true')
cg, errcg = np.loadtxt("ISING_1D/Gibbs_H_0.00/heat_capacity.out", usecols = (0,1), delimiter=' ', unpack ='true')
mm, errmm = np.loadtxt("ISING_1D/Metro_H_0.02/magnetization.out", usecols = (0,1), delimiter=' ', unpack ='true')
mg, errmg = np.loadtxt("ISING_1D/Gibbs_H_0.02/magnetization.out", usecols = (0,1), delimiter=' ', unpack ='true')
xm, errxm = np.loadtxt("ISING_1D/Metro_H_0.00/susceptibility.out", usecols = (0,1), delimiter=' ', unpack ='true')
xg, errxg = np.loadtxt("ISING_1D/Gibbs_H_0.00/susceptibility.out", usecols = (0,1), delimiter=' ', unpack ='true')

um = np.flip(um)
ug = np.flip(ug)
cm = np.flip(cm)
cg = np.flip(cg)
mm = np.flip(mm)
mg = np.flip(mg)
xm = np.flip(xm)
xg = np.flip(xg)
errum = np.flip(errum)
errug = np.flip(errug)
errcm = np.flip(errcm)
errcg = np.flip(errcg)
errmm = np.flip(errmm)
errmg = np.flip(errmg)
errxm = np.flip(errxm)
errxg = np.flip(errxg)


T = (T1-T0)*(np.arange(len(um))/(len(um)-1)) + 0.5

th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )

plt.figure(figsize=(9,6))
plt.plot(T, e, '--', color='black', label='Solution')
plt.errorbar(T, um, yerr=errum, color='blue', label='Metropolis')
plt.errorbar(T, ug, yerr=errug, color='red', label='Gibbs')
plt.xlabel('T', fontsize=18)
plt.ylabel('$U(N,T)/N$', fontsize=18)
plt.title('Energy per Spin', fontsize=22)
plt.grid(True)
plt.legend()
plt.show()

heat=Ns*((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

plt.figure(figsize=(9,6))
plt.plot(T, heat, '--', color='black', label='Solution')
plt.errorbar(T, cm, yerr=errcm, color='blue', label='Metropolis')
plt.errorbar(T, cg, yerr=errcg, color='red', label='Gibbs')
plt.xlabel('T', fontsize=18)
plt.ylabel('$C(N,T)$', fontsize=18)
plt.title('Heat Capacity', fontsize=22)
plt.grid(True)
plt.legend()
plt.show()

h=0.02 #external field
b = 1/T
l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)


plt.figure(figsize=(9,6))
plt.plot(T, M, '--', color='black', label='Solution')
plt.errorbar(T, mm, yerr=errmm, color='blue', label='Metropolis')
plt.errorbar(T, mg, yerr=errmg, color='red', label='Gibbs')
plt.xlabel('T', fontsize=18)
plt.ylabel('$M(N,T)/N$', fontsize=18)
plt.title('Magnetization per Spin', fontsize=22)
plt.grid(True)
plt.legend()
plt.show()

X = Ns*beta*np.exp(2*beta*J)*(1-thN)/(1+thN)

plt.figure(figsize=(9,6))
plt.plot(T, X, color='black', label='Solution')
plt.errorbar(T, xm, yerr=errxm, color='blue', label='Metropolis')
plt.errorbar(T, xg, yerr=errxg, color='red', label='Gibbs')
plt.xlabel('T', fontsize=18)
plt.ylabel(r'$\chi(N,T)$', fontsize=18)
plt.title('Susceptibility', fontsize=22)
plt.grid(True)
plt.legend()
plt.show()
```

Effettivamente si vede un buon accordo fra i valori simulati con entrambi e i metodi e la soluzione analaitica. Di seguito sono rappresentati i grafici delle differenze fra i valori otteniìuti con i campionamenti e le soluzioni analitiche.

```python
plt.figure(figsize=(9,6))
plt.errorbar(T, um-e, yerr=errum, color='blue', label='Metropolis')
plt.errorbar(T, ug-e, yerr=errug, color='red', label='Gibbs')
plt.xlabel('T', fontsize=18)
plt.ylabel('$\Delta U(N,T)/N$', fontsize=18)
plt.title('Energy per Spin', fontsize=22)
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(9,6))
plt.errorbar(T, cm -heat, yerr=errcm, color='blue', label='Metropolis')
plt.errorbar(T, cg- heat, yerr=errcg, color='red', label='Gibbs')
plt.xlabel('T', fontsize=18)
plt.ylabel('$\Delta C(N,T)$', fontsize=18)
plt.title('Heat Capacity', fontsize=22)
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(9,6))
plt.errorbar(T, mm - M, yerr=errmm, color='blue', label='Metropolis')
plt.errorbar(T, mg - M, yerr=errmg, color='red', label='Gibbs')
plt.xlabel('T', fontsize=18)
plt.ylabel('$\Delta M(N,T)/N$', fontsize=18)
plt.title('Magnetization per Spin', fontsize=22)
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(9,6))
plt.errorbar(T, xm - X, yerr=errxm, color='blue', label='Metropolis')
plt.errorbar(T, xg - X, yerr=errxg, color='red', label='Gibbs')
plt.xlabel('T', fontsize=18)
plt.ylabel(r'$\Delta \chi(N,T)$', fontsize=18)
plt.title('Susceptibility', fontsize=22)
plt.grid(True)
plt.legend()
plt.show()
```

Si vede come in generale, con i due algoritmi utilizzati sia più difficile simulare correttamente delle temperature più basse rispetto a quelle più alte. Alle temperature più basse sono probabilmente necessari molti più step per arrivare ad avere dei valori più precisi con incertezze minori.
