### Esercizio 03.1

In questo esercizio vogliamo verificare  numericamente la soluzione analitica della teoria di Black-Scholes. In particola tramite simulazione numerica vogliamo stimare qual è il *fair price* delle Opzioni Europee di tipo  *Call* e *Put*.

La teoria di Black-Scholes stabilisce che il prezzo segue un Moto Browniano Geometrico che ha come *drift* il Tasso d'interesse Risk-free $r$ e volatilità $\sigma$.

La soluzione analitica per il Moto Browniano $S\sim GBM(\mu,\sigma^2)$ con valore iniziale $S(0)$,  è data da:

$$S(t) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)t+\sigma W(t)\right]$$

dove $W(t)\sim\mathcal{N}(0,t)$.

Una proprietà importante del moto Browniano Geometrico è che anche nel caso di discretizzazione non ci sono errori, poichè gli incrementi di $W$ sono statisticamente indipendenti e normalmente distribuiti. Per cui possiamo anche valutare il prezzo al tempo t dividendo l'intervallo $[0,t]$ in n sottointevalli evalutando il valore di $S$ ricorsivamente
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma Z_{i+1}\sqrt{t_{i+1}-t_i}\right]$$
    con $Z_i \sim \mathcal{N}(0,1)$
</p>

La simulazione avviene campionando $M = 10^5$ volte il moto browniano, sia nel caso di campionamento diretto che nel caso di discretizzazione. Il prezzo equo (*fair price*) viene poi stimato come:
- Call:
$$ C = e^{-rT} \big<\big( S(T) - K \big)^+\big>$$
- Put:
$$ P = e^{-rT} \big<\big( K - S(T) \big)^+\big>$$

dove $(x)^+ = \max\{0,x\}$. <br>
La stima dell'errore invece avviene tramite il *Data Blocking* dividendo i $10^5$  campioni in $N=100$ blocchi.

I valori dei parametri utlizzati in questo problema sono:
- Prezzo $t=0$: $S(0)=100$
- Tempo di consegna: $T=1$
- Prezzo pattuito: $K=100$
- Tasso di interesse risk-free: $r=0.1$
- Volatilità: $\sigma=0.25$

L'obiettivo principale del codice è quello di implementare il Moto Browniano Geometrico. Questo viene fatto con la classe **BrownianMotion** dichiarata in *brownianMtion.h*. I metodi principali sono **BW** che implementa il Moto Browniano standard e soprattutto **GBW** che impplementa il Moto Browniano Geometrico. Nel caso di **GBW** bisogna specificare come dati di input il valore attuale del prezzo **S** e il tempo di evoluazione **t** del Moto Browniano Geometrico. Nel caso di campionamento diretto, t corrisponde al tempo di consegna e S al prezzo iniziale. Nel campionamento discretizzato invece t è il valore di ogni sottointervallo di tempo in cui è stato diviso il tempo di consegna T mentre S assume i diversi valori di prezzo ad ogni step.

I quattro grafici sottostanti rappresentano le l'andamento delle stime del prezzo delle Opzioni europee Call e Put, nel caso di campionamento diretto e discretizzato, in funzione del numero dei blocchi. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rc('text', usetex=True)
plt.rc('font', family='arial')

val, err = np.loadtxt("direct_call.dat", usecols=(0,1), delimiter=' ', unpack='true')
x = np.arange(len(val)) + 2

plt.figure(figsize=(9,6))
plt.errorbar(x,val,yerr=err, color='green')
plt.xlabel(r'Number of Blocks', fontsize=18)
plt.ylabel(r'$C$', fontsize=18)
plt.grid(True)
plt.title('Call - Direct sampling', fontsize=22)

plt.show()

print("Estimate: " + str(val[len(val)-1]))
print("Error: " + str(err[len(val)-1]))
```

```python
val, err = np.loadtxt("direct_put.dat", usecols=(0,1), delimiter=' ', unpack='true')
x = np.arange(len(val)) + 2

plt.figure(figsize=(9,6))
plt.errorbar(x,val,yerr=err, color='green')
plt.xlabel(r'Number of Blocks', fontsize=18)
plt.ylabel(r'$P$', fontsize=18)
plt.grid(True)
plt.title('Put - Direct sampling', fontsize=22)

plt.show()

print("Estimate: " + str(val[len(val)-1]))
print("Error: " + str(err[len(val)-1]))
```

```python
val, err = np.loadtxt("discrete_call.dat", usecols=(0,1), delimiter=' ', unpack='true')
x = np.arange(len(val)) + 2

plt.figure(figsize=(9,6))
plt.errorbar(x,val,yerr=err, color='green')
plt.xlabel(r'Number of Blocks', fontsize=18)
plt.ylabel(r'$C$', fontsize=18)
plt.grid(True)
plt.title('Call - Discretized sampling', fontsize=22)

plt.show()

print("Estimate: " + str(val[len(val)-1]))
print("Error: " + str(err[len(val)-1]))
```

```python
val, err = np.loadtxt("discrete_put.dat", usecols=(0,1), delimiter=' ', unpack='true')
x = np.arange(len(val)) + 2

plt.figure(figsize=(9,6))
plt.errorbar(x,val,yerr=err, color='green')
plt.xlabel(r'Number of Blocks', fontsize=18)
plt.ylabel(r'$P$', fontsize=18)
plt.grid(True)
plt.title('Put - Discrete sampling', fontsize=22)

plt.show()

print("Estimate: " + str(val[len(val)-1]))
print("Error: " + str(err[len(val)-1]))
```

La soluzione analitica di Black-Scholes da come risultati:

In tutti e quattro i casi le stime ottenute e le relative incertezze sono compatibili con la soluzione analitica.
